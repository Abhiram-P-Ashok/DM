```python
%pip install scikit-learn-extra
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn_extra.cluster import KMedoids
import matplotlib.pyplot as plt
```

```python
# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target
```

```python
# Perform K-medoids clustering
kmedoids = KMedoids(n_clusters=3, random_state=42)
kmedoids.fit(X)
cluster_labels = kmedoids.labels_
```

```python
# Plot the clustered data
plt.figure(figsize=(10, 6))
# Plot each cluster
for cluster in np.unique(cluster_labels):
    plt.scatter(X[cluster_labels == cluster, 0], X[cluster_labels == cluster, 1], label=f'Cluster {cluster}')
# Plot the centroids
centers = kmedoids.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], s=200, c='red', label='Centroids', marker='X')

plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('K-medoids Clustering of Iris Dataset')
plt.legend()
plt.show()
```
